```python
from tagger.core.mongo.models.notification import Notification, NotificationLog
```

```python
for n in Notification.objects():
    if NotificationLog.objects(notificationid=n.id).count() > 0:
        continue
    log = NotificationLog(
        result=n.result,
        notificationid=n.id,
        sent=n.sended
    )
    log.save()
    print(f"{n.id} -> {log.id}")
```

```python
for log in NotificationLog.objects():
    noti = Notification.objects(id=log.notificationid).first()
    if noti is None:
        continue
    noti.result = None
    noti.save()
```
